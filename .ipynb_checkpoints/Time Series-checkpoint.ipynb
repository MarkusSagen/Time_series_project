````python
import pandas as pd
import pandas_datareader as pdr
import statsmodels.tsa
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima_model import arma2ma

# Lets fo some ETS decomposition
airline = pd.read_csv('csv_data/airline_passengers.csv',index_col='Month')
airline.dropna(inplace=True)
airline.plot()
````

````python
# " Simple moving average "

airline.index
airline['6-Month-SMA']  = airline['Thousands of Passengers'].rolling(window=6).mean()
airline['12-Month-SMA'] = airline['Thousands of Passengers'].rolling(window=12).mean()
airline.plot(figsize=(12,4))
````

````python
# " Holt Winters method "
import pandas as pd
import pandas_datareader as pdr
import statsmodels.tsa
from statsmodels.tsa.holtwinters import SimpleExpSmoothing

# Lets fo some ETS decomposition
airline = pd.read_csv('csv_data/airline_passengers.csv',index_col='Month')
airline.index.freq = 'MS'     # Monthly start
airline.dropna(inplace=True)
airline.plot()
````

````python
# "" Simple Exp. Smoothing "
span = 12
alpha = 2/(span+1)

airline['EWMA12'] = airline['Thousands of Passengers'].ewm(alpha=alpha,adjust=False).mean()
airline.head()
````

````python
airline['SES12'] = SimpleExpSmoothing(airline['Thousands of Passengers']).fit(smoothing_level=alpha,optimized=False).fittedvalues.shift(-1)
airline.head()
airline.plot()
````

````python
# Double exp. smoothing
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Addative - Linear model
airline['DES_add_12'] = ExponentialSmoothing(airline['Thousands of Passengers'],trend='add').fit().fittedvalues.shift(-1)
airline[['Thousands of Passengers','SES12','DES_add_12']].iloc[:24].plot(figsize=(12,5))
airline[['Thousands of Passengers','SES12','DES_add_12']].iloc[-24:].plot(figsize=(12,5))
````

````python
# Double exp. smoothing
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Multiplicative - Exponential - Fitts really well!
airline['DES_mul_12'] = ExponentialSmoothing(airline['Thousands of Passengers'],trend='mul').fit().fittedvalues.shift(-1)
airline[['Thousands of Passengers','SES12','DES_add_12','DES_mul_12']].iloc[:24].plot(figsize=(12,5))
airline[['Thousands of Passengers','SES12','DES_add_12','DES_mul_12']].iloc[-24:].plot(figsize=(12,5))
````

````python
# Tripple exp. smoothing
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Multiplicative - Exponential - Fitts really well!
airline['TES_mul_12'] = ExponentialSmoothing(airline['Thousands of Passengers'],trend='mul',seasonal='mul',seasonal_periods=12).fit().fittedvalues
airline[['Thousands of Passengers','DES_mul_12','TES_mul_12']].iloc[:24].plot(figsize=(12,5))
airline[['Thousands of Passengers','DES_mul_12','TES_mul_12']].iloc[-24:].plot(figsize=(12,5))
````

````python
# " Holt Winters method "
import pandas as pd
import pandas_datareader as pdr
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Lets fo some ETS decomposition
airline = pd.read_csv('csv_data/airline_passengers.csv',index_col='Month',parse_dates=True)
airline.index.freq = 'MS'     # Monthly start
airline.dropna(inplace=True)

airline.info()  # 144 entries - Idealy split 80/20
train_data = airline.iloc[:109]
test_data = airline.iloc[108:]
````

# Descriptive statistics test

__Dickey-Fuller Test__ A test for testing stationarity    
We start to assume that \theta = 1 as the null hypothesis  
  
IF p < 0.005, (the probability) then we reject the null hypothesis  
IF p > 0.05, we fail to reject null hypothesis  
  
__Granger Causality Test__  
Used to see indications of inside causality, but it does **NOT** account for outside factors   
  
We have used MSE,MAE,RMSE  

AIC - Akaike Information Criteria - One of the moste cited papers   
Estimates  
   
BIC - Bayesian Information Criteria - Simular to AIC, but uses Bayesian statistics

````python
from statsmodels.tsa.arima_model import ARMA,ARIMA,ARMAResults,ARIMAResults
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
from pmdarima import auto_arima

df1 = pd.read_csv('csv_data/DailyTotalFemaleBirths.csv',index_col='Date',parse_dates=True)
df1.index.freq = 'D'
df1 = df1[:120]

df2 = pd.read_csv('csv_data/TradeInventories.csv',index_col='Date',parse_dates=True)
df2.index.freq = 'MS'

df1.plot(figsize=(12,5)) # Assume Stationary, lets check
````

````python
# Import check for stationary


from statsmodels.tsa.stattools import adfuller

def adf_test(series,title=''):
    """ Pass in time series and recive the ADF report """
    print(f'Augmented Dickey-Fuller Test: {title}')
    result = adfuller(series.dropna(),autolag='AIC')
    
    labels = ['ADF test statistics','p-value','# lags used','# observations']
    out = pd.Series(result[0:4],index=labels)
    
    for key,value in result[4].items():
        out[f'critical value ({key})'] = value
        
    print(out.to_string())
    
    # Check null hypothesis
    if result[1] <= 0.05:
        print("Reject Null Hypothesis")
        print("Data has no unit root and is stationary") 
    else: 
        print("Fail to reject Null Hypothesis")
        print("Data has a unit root and is non-stationary")
````
